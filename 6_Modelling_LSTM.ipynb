```python
# Import Libraries

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
import warnings
warnings.filterwarnings("ignore")

# LSTM model libraries
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping
from keras.regularizers import L1L2
from tensorflow.keras.layers import Activation, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import callbacks
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
from tensorflow.keras import backend as K
```

### Import data

```python
final_df = pd.read_csv('final_df_next_7th_day_15.csv').reset_index(drop=True)
pd.set_option('display.float_format', lambda x: '%.9f' % x)
final_df.head()
```

```python
def calculate_metrics(y, ypred):
    MAE = metrics.mean_absolute_error(y, ypred)
    MSE = metrics.mean_squared_error(y, ypred,squared=True)
    RMSE = metrics.mean_squared_error(y, ypred,squared=False)
    R2 = metrics.r2_score(y, ypred)
    return MAE,MSE,RMSE,R2

def plot_result(df,title=None):
    df_copy = df.copy()
    
    # Convert 'Date' column to datetime format
    df_copy['Date'] = pd.to_datetime(df_copy['Date'])
    
    # Set the 'Date' column as index
    df_copy.set_index('Date', inplace=True)

    # Create the plot
    plt.figure(figsize=(25, 6))
    plt.plot(df_copy.iloc[:, 0], label='Actual')
    plt.plot(df_copy.iloc[:, 1], label='Predicted')

    # Label the axes
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.title(title)
    plt.show()
```

### LSTM

```python
K.clear_session()
tf.compat.v1.reset_default_graph()

# Determine the size of the train and test set
test_size = 0.3
n_test = int(np.round(len(final_df) * test_size))
train_split = final_df[:len(final_df)-n_test]
test_split = final_df[len(final_df)-n_test:]

x_train = train_split.drop(['next_7th_day_closing_price','Date'],axis=1).values
x_test = test_split.drop(['next_7th_day_closing_price','Date'],axis=1).values

y_train = train_split['next_7th_day_closing_price'].reset_index(drop=True).values
y_test = test_split['next_7th_day_closing_price'].reset_index(drop=True).values

# reshape input to be [samples, time steps, features] which is required for LSTM
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1] , 1)
x_test = x_test.reshape(x_test.shape[0],x_test.shape[1] , 1)


model = Sequential()

# Adding a LSTM layer with 100 internal units, L1L2 regularization, and dropout
model.add(LSTM(100, input_shape=(None, 1), activation='relu', kernel_regularizer=L1L2(l1=0.001, l2=0.001), recurrent_dropout=0.2))

# Adding a Dense layer with 1 unit
model.add(Dense(1))

# Define a custom learning rate
learning_rate = 0.001
optimizer = Adam(lr=learning_rate)

# Loss function + optimizer
model.compile(loss='huber_loss', optimizer=optimizer)

# Add early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Train the model
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=32, verbose=1, callbacks=[early_stopping])
y_train_predict=model.predict(x_train)
y_test_predict=model.predict(x_test)

MAE_train,MSE_train,RMSE_train,R2_train = calculate_metrics(y_train,y_train_predict)
MAE_test,MSE_test,RMSE_test,R2_test = calculate_metrics(y_test,y_test_predict)

lstm_result_test_df = pd.DataFrame()
lstm_result_test_df['Date'] = test_split['Date']
lstm_result_test_df['y_test'] = test_split['next_7th_day_closing_price']
lstm_result_test_df['y_test_pred'] = y_test_predict.flatten()


lstm_result_train_df = pd.DataFrame()
lstm_result_train_df['Date'] = train_split['Date']
lstm_result_train_df['y_train'] = train_split['next_7th_day_closing_price']
lstm_result_train_df['y_train_pred'] = y_train_predict.flatten()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(loss))
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend(loc=0)
plt.figure()
plt.show()
```

```python
print("Mean Absolute Error (MAE) for Training:", MAE_train)
print("Mean Squared Error (MSE) for Training:", MSE_train)
print("Root Mean Squared Error (RMSE) for Training:", RMSE_train)
print("R-squared score (R2) for Training:", R2_train)
print("********************************************")
print("Mean Absolute Error (MAE) for Testing:", MAE_test)
print("Mean Squared Error (MSE) for Testing:", MSE_test)
print("Root Mean Squared Error (RMSE) for Testing:", RMSE_test)
print("R-squared score (R2) for Testing:", R2_test)
```

```python
plot_result(lstm_result_test_df,'results-testing')
```

```python
plot_result(lstm_result_train_df,'results-training')
```
